```python
'''import all necessary libraries'''

import cx_Oracle #import the oracle database libraries
import pandas as pd #connect to the pandas libraries
import numpy as np #import numpy
import getpass #import the password libraries in order to keep database password concealed

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
import sys
print (sys.version)
pd.options.display.float_format = '{:0.6f}'.format
```

```python
'''connect to the oracle server used to pull down the 
currency and misuse of authorization information'''

username=input("please enter your username")
password=getpass.getpass("Enter your password: ")

con=cx_Oracle.connect(username,password,'ODSPEU1.ams') #enter database details
print (con.version) # prnt the conversion details
```

```python
'''client names'''
df_cn=pd.read_sql(
''' 
select MM.CONTRACT_ID, MM.CLIENT_NAME
from FDWO.MB_MERCHANTS mm
group by MM.CONTRACT_ID, MM.CLIENT_NAME
''', con)
```

```python
df_cn.info()
```

```python
columns=['MONTH','TRADING_MERCHANT_NAME','CONTRACT_ID','ISSUER_COUNTRY','MERCHANT_COUNTRY'
         ,'ISSUER_COUNTRY_VISA_EUR_REL','ISSUER_COUNTRY_MC_SEPA_REL','INTERCHANGE_RATE_DESCRIPTION'
         ,'CARD_BRAND','SECURITY_LEVEL','FLOW','VOL','SF','SF_RATE_AMOUNT','PRODUCT_CODE','TRANSACTION_CURRENCY','TRANSACTION_FUNDING_CURRENCY'
         ,'SF_RATE_PERC','SF_RATE_DESCRIPTION','CARD_TYPE_GROUP','CARD_TYPE'
         ,'Security Level','Regionality','BRAND','CARD_TYPES']

df1=pd.read_csv('u:\\aib_theoretical_dataset.txt',encoding='ISO-8859-1',sep=';',skiprows=39,low_memory=False,names=columns)
```

```python
df2=pd.read_csv('d:\\djohnson\\Desktop\\aib_hypotheticals\\aib_iep_theoreticals_V3.csv',encoding='ISO-8859-1')
df_ic=pd.read_csv('d:\\djohnson\\Desktop\\aib_hypotheticals\\ic_plus_type.csv',encoding='ISO-8859-1')
df_mc=pd.read_csv('d:\\djohnson\\Desktop\\aib_hypotheticals\\multicurrency.csv',encoding='ISO-8859-1')
```

```python
df=df1.copy()
df=df.fillna('').copy()
df.head()
```

```python
df_ic=df_ic.drop(['KEY','MATCH'],axis=1)
df_ic.info()
```

```python
df['TRANSACTION_CURRENCY']=df['TRANSACTION_CURRENCY'].str.strip()
df['TRANSACTION_FUNDING_CURRENCY']=df['TRANSACTION_FUNDING_CURRENCY'].str.strip()
df['SF_RATE_PERC']=df['SF_RATE_PERC'].str.strip()
df['PRODUCT_CODE']=df['PRODUCT_CODE'].str.strip()
df['SF_RATE_AMOUNT']=df['SF_RATE_AMOUNT'].str.strip()
df['TRADING_MERCHANT_NAME']=df['TRADING_MERCHANT_NAME'].str.strip()
df['ISSUER_COUNTRY']=df['ISSUER_COUNTRY'].str.strip()
df['MERCHANT_COUNTRY']=df['MERCHANT_COUNTRY'].str.strip()
df['ISSUER_COUNTRY_VISA_EUR_REL']=df['ISSUER_COUNTRY_VISA_EUR_REL'].str.strip()
df['ISSUER_COUNTRY_MC_SEPA_REL']=df['ISSUER_COUNTRY_MC_SEPA_REL'].str.strip()
df['INTERCHANGE_RATE_DESCRIPTION']=df['INTERCHANGE_RATE_DESCRIPTION'].str.strip()
df['CARD_BRAND']=df['CARD_BRAND'].str.strip()
df['SECURITY_LEVEL']=df['SECURITY_LEVEL'].str.strip()
df['SF_RATE_DESCRIPTION']=df['SF_RATE_DESCRIPTION'].str.strip()
df['CARD_TYPE_GROUP']=df['CARD_TYPE_GROUP'].str.strip()
df['CARD_TYPE']=df['CARD_TYPE'].str.strip()
df['Security Level']=df['Security Level'].str.strip()
df['Regionality']=df['Regionality'].str.strip()
df['BRAND']=df['BRAND'].str.strip()
df['CARD_TYPES']=df['CARD_TYPES'].str.strip()
df['CONTRACT_ID']=df['CONTRACT_ID'].str.strip()
```

```python
'''combine client id information with main dataset'''
df=pd.merge(df,df_cn,on=['CONTRACT_ID'],how='left')

'''build period of time which will be used to determine which fees will be applied'''
df['PERIOD']=''
df['PERIOD'][(df['MONTH']<7)]='prejuly18'
df['PERIOD'][(df['MONTH']>=7)]='postjuly18'
df.info()
```

```python
df=df[(df['SF_RATE_AMOUNT']!='')].copy()
df=df[(df['PRODUCT_CODE']!='')].copy()
df=df[(df['CONTRACT_ID']!='')].copy()
df['SF_RATE_AMOUNT']=df['SF_RATE_AMOUNT'].astype(float)
df['SF_RATE_PERC']=df['SF_RATE_PERC'].astype(float)
df['PRODUCT_CODE']=df['PRODUCT_CODE'].astype(float)
df['CONTRACT_ID']=df['CONTRACT_ID'].astype(int)
```

```python
df['Regionality'][(df['ISSUER_COUNTRY_MC_SEPA_REL']=='SEPA')&(df['CARD_BRAND']=='MasterCard')&(df['Regionality']=='Intra-Region')]='Inter-Region'
df['Regionality'][(df['ISSUER_COUNTRY_MC_SEPA_REL']=='SEPA')&(df['CARD_BRAND']=='Maestro')&(df['Regionality']=='Intra-Region')]='Inter-Region'
```

```python
'''Adjust the regionality to ensure that certain countries are being assessed the correct scheme fees'''

df=df[(df['Regionality']!='')].copy()
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='United Kingdom')]='UK'

df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Russian Federation')&(df['CARD_BRAND']=='MasterCard')]='Russia'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Russian Federation')&(df['CARD_BRAND']=='Maestro')]='Russia'

df['MERCHANT_COUNTRY_UPDATED_FROM:']=''
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='Gibraltar')]='Gibraltar'
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='Andorra')]='Andorra'
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='Panama')]='Panama'
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='Monaco')]='Monaco'
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='Australia')]='Australia'
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='New Zealand')]='New Zealand'
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='Singapore')]='Singapore'
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='Thailand')]='Thailand'
df['MERCHANT_COUNTRY_UPDATED_FROM:'][(df['MERCHANT_COUNTRY']=='Malaysia')]='Malaysia'

df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Gibraltar')]='UK'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Andorra')]='France'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Panama')]='Ireland'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Monaco')]='France'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Australia')]='Ireland'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='New Zealand')]='Ireland'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Singapore')]='Ireland'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Thailand')]='Ireland'
df['MERCHANT_COUNTRY'][(df['MERCHANT_COUNTRY']=='Malaysia')]='Ireland'

df['Product']=df['BRAND']+df['CARD_TYPES']
df['Concatenate']=df['Product']+df['MERCHANT_COUNTRY']+df['Regionality']
df.head()
```

```python
df2=df2[['Concatenate','Security Level','Fixed','Percent','PERIOD']]
df2['Percent']=df2['Percent'].astype(float)
```

```python
df3=pd.merge(df,df2,on=['Security Level','Concatenate','PERIOD'],how='left')
df3['Fixed']=df3['Fixed'].astype(float)
df3['FLOW']=df3['FLOW'].astype(float)
df3['VOL']=df3['VOL'].astype(float)
```

```python
df3=pd.merge(df3,df_mc,on=['MERCHANT_COUNTRY'],how='left').copy()
```

```python
df3.info()
```

```python
df3['MULTI_CURRENCY_FEE']=0
```

```python
df3['MULTI_CURRENCY_FEE'][(df3['MULTI_CURRENCY_APPLICABLE']=='Yes')&(df3['BRAND']=='MasterCard')&(df3['Regionality']=='Domestic')&(df3['TRANSACTION_CURRENCY']!=df3['TRANSACTION_FUNDING_CURRENCY'])]=.0003
df3['MULTI_CURRENCY_FEE'][(df3['MULTI_CURRENCY_APPLICABLE']=='Yes')&(df3['BRAND']=='Maestro')&(df3['Regionality']=='Domestic')&(df3['TRANSACTION_CURRENCY']!=df3['TRANSACTION_FUNDING_CURRENCY'])]=.0003

df3['MULTI_CURRENCY_FEE'][(df3['MULTI_CURRENCY_APPLICABLE']=='Yes')&(df3['BRAND']=='MasterCard')&(df3['Regionality']=='Intra-Region')&(df3['TRANSACTION_CURRENCY']!=df3['TRANSACTION_FUNDING_CURRENCY'])]=.0003
df3['MULTI_CURRENCY_FEE'][(df3['MULTI_CURRENCY_APPLICABLE']=='Yes')&(df3['BRAND']=='Maestro')&(df3['Regionality']=='Intra-Region')&(df3['TRANSACTION_CURRENCY']!=df3['TRANSACTION_FUNDING_CURRENCY'])]=.0003

df3['MULTI_CURRENCY_FEE'][(df3['MULTI_CURRENCY_APPLICABLE']=='Yes')&(df3['BRAND']=='MasterCard')&(df3['Regionality']=='Inter-Region')&(df3['TRANSACTION_CURRENCY']!=df3['TRANSACTION_FUNDING_CURRENCY'])]=.0013
df3['MULTI_CURRENCY_FEE'][(df3['MULTI_CURRENCY_APPLICABLE']=='Yes')&(df3['BRAND']=='Maestro')&(df3['Regionality']=='Inter-Region')&(df3['TRANSACTION_CURRENCY']!=df3['TRANSACTION_FUNDING_CURRENCY'])]=.0013
```

```python
df3['HYPOTHETICAL_SF']=(df3['Fixed']*df3['VOL'])+(df3['Percent']*df3['FLOW'])+(df3['MULTI_CURRENCY_FEE']*df3['FLOW'])
```
