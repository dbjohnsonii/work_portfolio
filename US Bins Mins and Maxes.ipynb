In our portfolio there are multiple types of cards. For example there are MasterCard Credit Consumer Classic Cards or Visa Credit Consumer Classic Cards. These descriptions howerver can be broad. Depending under which BIN these broad descriptions fall costs can range.

What my manager requested in this exercise was that I pull a set of information and under the broad descriptions identify the min and maxes of each category and assemble them in order to provide a merchant with a range for each category their customers were using.

```python
#import pandas and the cost information from the local drive
import pandas as pd
df=pd.read_csv('d:\\djohnson\\Desktop\\US BINS MINS AND MAXES.csv')
```

```python
#setup empty lists for the information to flow into
key=[]
min_ic_perc=[]
min_ic_rate=[]
max_ic_perc=[]
max_ic_rate=[]
```

```python
for x in range(len(df.KEY.unique())): #the dataset has keys built in what I'm asking the script  
    selection=df.KEY.unique()[x]      #to do here is find each unique key and to provide a count    
    key.append(selection)             #after the unique categories are identified the results based on their number in the 
                                      #sequence is appended to a list              
    
    df_min=df[(df['KEY']==selection)] #the selection will then seek the entire list of the category and isolate all instances
    df_min=df_min[(df_min['$COST']==df_min['$COST'].min())] #from all of the instances the script asks to be shown the instance with the lowest cost and establishes the lowest
    df_min=df_min[['KEY','INTERCHANGE_PERCENT_RATE','INTERCHANGE_FLAT_RATE']] #the result is then parsed between the flat and percentage
    df_min=df_min[0:1]
    min_ic_perc.append(df_min.iloc[0,1]) #and each is appended into the rate or percentage list. This then create lists that correspond with that key
    min_ic_rate.append(df_min.iloc[0,2])
    
    df_max=df[(df['KEY']==selection)] #the same process is then executed for maxes
    df_max=df_max[(df_max['$COST']==df_max['$COST'].max())]
    df_max=df_max[['KEY','INTERCHANGE_PERCENT_RATE','INTERCHANGE_FLAT_RATE']]
    df_max=df_max[0:1]
    max_ic_perc.append(df_max.iloc[0,1])
    max_ic_rate.append(df_max.iloc[0,2])
```

```python
#after the creation of the lists, the information is converted to dataframes...
key=pd.DataFrame(key,columns=['key'])
min_ic_perc=pd.DataFrame(min_ic_perc,columns=['min_ic_perc'])
min_ic_rate=pd.DataFrame(min_ic_rate,columns=['min_ic_rate'])
max_ic_perc=pd.DataFrame(max_ic_perc,columns=['max_ic_perc'])
max_ic_rate=pd.DataFrame(max_ic_rate,columns=['max_ic_rate'])
#whcih can be easily joined together...
output=key.join(min_ic_perc).join(min_ic_rate).join(max_ic_perc).join(max_ic_rate)
```

```python
#the output is a comprehensive dataframe my manager can take back to the merchant to show under each category what min and max costs
#they are paying
output
```
